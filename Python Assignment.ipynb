## III.	Determine the accuracy of the Coordinates with Address.  Add another column GeoAccuracy.  Find the coordinates using the address.  If the coordinates match exactly with address, then accuracy is 100%, otherwise for every 10 meters off consider a drop in accuracy of 1 %.  For example, if the coordinate of the address is 30 meters away from the given coordinates, the accuracy would be 70%.

```python
def get_estimated_accuracy(address, coordinates):
    """
    Estimates the accuracy of coordinates based on their format and known precisions.

    **Note:** This is a simplified approach without external libraries and may not be very accurate.

    Args:
        address (str): The address to check.
        coordinates (tuple): A tuple containing latitude and longitude.

    Returns:
        int: Estimated accuracy as a percentage (between 0 and 100).
    """

    accuracy = 0

    # Check address format
    if any(char.isalpha() for char in address if char not in " "):  # Address contains letters, likely a street name
        accuracy += 20

    # Check coordinate precision (assuming more decimal places = higher accuracy)
    lat_precision = len(str(coordinates).split('.')[1]) if len(str(coordinates).split('.')) > 1 else 0  # Access the entire tuple as a string

    accuracy += min(lat_precision, 20)  # Limit increase to 20%

    return accuracy
```

```python
delhi_electrict_substation_data['GeoAccuracy'] = delhi_electrict_substation_data.apply(
    lambda row: get_estimated_accuracy(row['Address'], row['Coordinates']), axis=1)
delhi_electrict_substation_data
```

## Q2.  The file MH-Veh-Reg.csv contains data of vehicles registered in Maharashtra for the period 2000-2018.   Write python code to do the following:

## i.	Merge the data of each year (2000-2001, 2001-2002,.) to a single period (2000-2018) by the region and sub region.

```python
data = pd.read_csv("MH-Vehicle-Reg.csv", sep=';', header=0, index_col=0)
data.head()
```

```python
data.groupby(['Region', 'Sub Region']).sum()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_vehicle_classification(year, df):
    # Classify vehicles into categories based on the given criteria
    df['Category'] = 'Private'
    df.loc[(df['Motor Cycles'] > 0) | (df['Scooters'] > 0) | (df['Moped'] > 0) | (df['Cars'] > 0) | (df['Jeeps'] > 0) | (df['Stn. Wagons'] > 0) | (df['Private Service Vehicles'] > 0), 'Category'] = 'Private'
    df.loc[(df['Category'] != 'Private'), 'Category'] = 'Commercial'
    
    df['Subcategory'] = 'LMV'
    df.loc[(df['School Buses'] > 0) | (df['Trucks & Lorries'] > 0) | (df['Trailors'] > 0) | (df['Tractors'] > 0) | (df['Tanker'] > 0) | (df['Stage carriages'] > 0) | (df['Contract carriages /Mini Bus'] > 0), 'Subcategory'] = 'HMV'

    # Calculate total count and percentage of each category
    category_counts = df['Category'].value_counts()
    category_percentages = category_counts / category_counts.sum() * 100
    
    subcategory_counts = df['Subcategory'].value_counts()
    subcategory_percentages = subcategory_counts / subcategory_counts.sum() * 100
    
    # Plotting
    plt.figure(figsize=(15, 6))
    
    # Bar plot for counts
    plt.subplot(1, 2, 1)
    category_counts.plot(kind='bar', color=['blue', 'green'], alpha=0.7)
    plt.title(f'Private vs Commercial Vehicles - {year}')
    plt.xlabel('Vehicle Category')
    plt.ylabel('Count')
    
    # Bar plot for percentages
    plt.subplot(1, 2, 2)
    category_percentages.plot(kind='bar', color=['blue', 'green'], alpha=0.7)
    plt.title(f'Percentage of Private vs Commercial Vehicles - {year}')
    plt.xlabel('Vehicle Category')
    plt.ylabel('Percentage (%)')
    
    # Adding legend
    plt.legend(['Private', 'Commercial'])
    
    # Show plots
    plt.tight_layout()
    plt.show()
    
    # Plotting for Subcategory
    plt.figure(figsize=(15, 6))
    
    # Bar plot for counts
    plt.subplot(1, 2, 1)
    subcategory_counts.plot(kind='bar', color=['blue', 'green'], alpha=0.7)
    plt.title(f'LMV vs HMV Vehicles - {year}')
    plt.xlabel('Vehicle Subcategory')
    plt.ylabel('Count')
    
    # Bar plot for percentages
    plt.subplot(1, 2, 2)
    subcategory_percentages.plot(kind='bar', color=['blue', 'green'], alpha=0.7)
    plt.title(f'Percentage of LMV vs HMV Vehicles - {year}')
    plt.xlabel('Vehicle Subcategory')
    plt.ylabel('Percentage (%)')
    
    # Adding legend
    plt.legend(['LMV', 'HMV'])
    
    # Show plots
    plt.tight_layout()
    plt.show()
```
