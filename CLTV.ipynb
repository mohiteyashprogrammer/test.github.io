```python
data = data.drop(["Description"],axis=1)
```

```python
# Calculate Average Purchase Value
average_purchase_value = data['UnitPrice'].mean()
```

```python
# Calculate Average Purchase Frequency
average_purchase_frequency = data["CustomerID"].nunique() / data.shape[0]
```

```python
average_purchase_frequency
```

```python
# Calculate Customer Value
customer_value = average_purchase_value * average_purchase_frequency

# Calculate Average Customer Lifespan (for example, let's assume 3 years)
average_customer_lifespan = 3

# Calculate CLTV
CLTV = customer_value * average_customer_lifespan
```

```python
# Calculate Average Purchase Value
data['TotalRevenue'] = data['Quantity'] * data['UnitPrice']
```

```python
verage_purchase_value = data.groupby('CustomerID')['TotalRevenue'].mean()
```

```python
verage_purchase_value 
```

```python
# Calculate Purchase Frequency
purchase_frequency = data.groupby('CustomerID').size()
purchase_frequency
```

```python
# Merge Average Purchase Value and Purchase Frequency into one DataFrame
cltv_data = pd.DataFrame({'AvgPurchaseValue': average_purchase_value, 'PurchaseFrequency': purchase_frequency})
```

```python
cltv_data
```

```python
# Calculate CLTV
cltv_data['CLTV'] = cltv_data['AvgPurchaseValue'] * cltv_data['PurchaseFrequency']
```

```python
# Merge CLTV data back to the original DataFrame
purchase_data = pd.merge(data, cltv_data['CLTV'], on='CustomerID', how='left')
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
#numeric pipline
num_pipline = Pipeline(
    steps=[
        ("imputer",SimpleImputer(strategy="median")),
        ("scaler",StandardScaler())
    ]
)

# cato_pipline
cato_pipline = Pipeline(
    steps=[
        ("imputer",SimpleImputer(strategy="most_frequent")),
        ("onehote",OneHotEncoder(sparse=False,handle_unknown="ignore",drop="first")),
        ("scaler",StandardScaler(with_mean=False))
    ]
)

# create preprocessor object
preprocessor = ColumnTransformer([
    ("num_pipline",num_pipline,numerical_features),
    ("cato_pipline",cato_pipline,catigorical_features)])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# apply preprocessor object and transform data
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)
```

```python
X_test
```

```python
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
```

```python
# this function will eveluate the model
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
def model_evalution(test,predict):
    mse = mean_squared_error(test,predict)
    mae = mean_absolute_error(test,predict)
    rmse = np.sqrt(mse)
    r2score = r2_score(test,predict)
    return mse,mae,rmse,r2score
```

```python
forest = RandomForestRegressor(
    n_estimators=100,
    max_depth=8
    
)
```

```python
forest.fit(X_train,y_train)
```

```python
y_predict = forest.predict(X_test)
```

```python
model_evalution(y_test,y_predict)
```
